```python
headers={
    "User-Agent": "Chrome"  #requst will sended as chrome
}
URL="https://www.scrapethissite.com/pages/simple/"
res= requests.get(URL,headers=headers)
```

```python


if (res.status_code==200):
    print(res)
with open("data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
#beutiful soup
from bs4 import BeautifulSoup

with open("data1.html","r") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")

```

```python
soup.find("h1") # find the first occurence 
```

```python
all_h3=soup.find_all("h3") #give all the occurence
all_countries=[]

for h3 in all_h3:
    name=(h3.get_text(strip=True)) #give the content of all the tag check it
    #remove the spaces strip=true
    population=(h3.find_next("div").select("span.country-population")[0].get_text(strip=True)) #find the next class 
    all_countries.append([name,population])
```

```python

import pandas as pd
pd.DataFrame(all_countries,columns=["Name","Population"])
```
