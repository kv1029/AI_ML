```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Fetch data & store it
page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No more valid pages...")
        break

    with open(f"quotes{page_count}.html", "w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"Page {page_count} downloaded")

    page_count += 1
```

```python
# Now for extracting the data
authors=[]
page_count=1;
while True:
    with open(f"quotes{page_count}.html","r",encoding ="utf-8") as f:
        html_content=f.read()
        
    soup=BeautifulSoup(html_content,"lxml")
    all_quotes= soup.select("div.quote")
    
    for q in all_quotes:
        all_tags=[]
        for p in q.select(".tags .tag"):
            all_tags.append(p.get_text())
        if "life" in all_tags:
            quot=q.select_one("span.text").get_text()
            name= q.select_one("small.author").get_text()
            authors.append([quot,name])
    if(page_count==10):
        break;
    page_count+=1                                                                                                                                             
    
```

```python
ans= pd.DataFrame(authors)
```
