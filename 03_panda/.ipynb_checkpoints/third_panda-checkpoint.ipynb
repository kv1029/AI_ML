```python
# aqi dataset
df= pd.read_csv("globalAirQuality.csv")
df.describe()

```

```python
#selecting data in columns
df["city"]
df[["city","aqi"]]
```

```python
# slecting data by rows
df.loc[0]
df.loc[0:2] # start idx:  end idx (inclusive)
df.iloc[0:2]  # start idx:  end idx (exclusive)


df.loc[0,"aqi"]
```

```python
# filtering the data
df[(df["aqi"]>100) & (df["temperature"]>30)]

aqi_data=df[(df["aqi"]>100) & (df["temperature"]>30)][["city","aqi"]]
aqi_data
aqi_data.iloc[0] # can print from 0
aqi_data.loc[6] #could not print form index 0 print form where it start
```

```python
# query in pandas
# query return the copy of data not view
df.query("aqi>100 & temperature >30")[["city","aqi"]] # case sensitive column name
# doing query by variable
aqi_val=100
df.query("aqi>@aqi_val & temperature >30")
```

```python
#clening the data pf raw_csv file
df= pd.read_csv("raw_data.csv")
df.isnull() # return the null values 
df.isnull().sum() # count no of null values in column
df.dropna() # drop full row which is null
df.dropna(axis=1) # drop column values
df.fillna(0) # fill null values with the given one
df["age"].fillna(df["age"].mean())
df.ffill() # fill the next null value with the curent one
df.bfill() #opposite of ffill
```

```python
df
```

```python
#apply mathod using lambda
df2=df.copy()
df2["tax"]=df2["income"].apply(lambda x:"20%" if x>=60000 else "10%")

#map keyword 
gender_map = {"Male" : "M", "Female": "F", "Unknown" : "0"}
df2["gender"]= df2["gender"].map(gender_map)
```

```python
df2
```

```python
df2.rename(columns={"income":"salary"})
 # to changes the name of columns
# df2.rename(index={1:"first"}) # to change the label name in data

# df.sort_values("income")
# df.sort_values("income",ascending =False)
sorted_df=df2.sort_values(["income","age"])
```

```python
sorted_df.reset_index()
#ranking
sorted_df["ranking"]=sorted_df["income"].rank()
sorted_df
#reorders columns
# Ranking
# sorted_df ["Ranking"] = sorted_df ["Income").rank(ascending=False, method="max"]
sorted_df[["id", "name", "age", "country", "gender", "income","tax"]]
sorted_df
```

```python
## shift the column row to the last of the column
df2 =df.copy()
new_col_order= [col for col in df2.columns if col != "id"]+["id"]
print(new_col_order)
df2 [new_col_order]
```

```python
# to write data in csv file or json
df2=df.copy()
df2=df2.drop_duplicates()
df2=df2.fillna(0)
df2=df2.sort_values("income")
df2=df2.reset_index(drop=True)
```

```python
df2.to_csv("sorted_data.csv")
```

```python
# grouping and aggregation 
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()


df.groupby("gender")["income"].mean()
df.groupby("country")["income"].agg(["mean","max","min"])
df.groupby("country").agg({
"income": "mean",
"age": "mean"
})
```
