```python
# Oscar Winner Movies
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M
# Plot
plt.plot(years, oscar_revenue)
plt.title("oscar revenue")  #add title to the graph
plt.xlabel("years")
plt.ylabel("revenue")
```

```python
#Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 82, 185, 275] # in $M
plt.plot(years,oscar_revenue,label="Oscar Movies")
plt.plot(years,non_oscar_revenue,"o--,label="Non Oscar Movies")
plt.title("oscar movies vs Non oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.legend() ##help to label the data line ex in first plot label="Oscar Movies"
```
