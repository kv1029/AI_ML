```python
# Oscar Winner Movies
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M
# Plot
plt.plot(years, oscar_revenue)
plt.title("oscar revenue")  #add title to the graph
plt.xlabel("years")
plt.ylabel("revenue")
```

```python
#Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 82, 185, 275] # in $M
plt.plot(years,oscar_revenue,"d--b",label="Oscar Movies")
plt.plot(years,non_oscar_revenue,"o--r",label="Non Oscar Movies")
plt.title("oscar movies vs Non oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.legend() ##help to label the data line ex in first plot label="Oscar Movies"
```

```python
# Plot Methods
# 1. plt.plot(x, y) - Line plot
# 2. plt.bar(x, height) - Vertical bar chart
# 3. plt.barh(y, width) - Horizontal bar chart
# 4. plt.scatter(x, y) - Scatter plot
# 5. plt.pie(data) - Pie chart
# 6. plt.hist(data, bins) - Histogram
# 7. plt.boxplot(data) - Box plot
```

```python
##### bar graph 
oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]
plt.bar(years,oscar_revenue,color="green")
plt.bar(years, oscar_revenue, color="green")
plt.title("Revenue in each Year for Oscar movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
for i in range(len(years)):
    plt.text(years [i], oscar_revenue [i]+5, str(oscar_revenue [i]), ha="center")
```

```python
## box plot for multiple dataset
group1=np.random.normal(50,10,100)
group2=np.random.normal(60,5,100)
plt.boxplot([group1,group2],tick_labels=["group1","group2"])
plt.grid(True)
plt.show()
```

```python
# for stack plots
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]
plt.stackplot(days, direct, organic, social,labels=["Direct", "Organic", "Social"])
plt.title("Marketing data")
plt.xlabel("Days")
plt.ylabel("No Of customers")
```

```python
#### Sub Plots #####
x=[1,2,3,4,5,6,7,8,9]
y1=[np.sqrt(i) for i in x]
y2=[i*2 for i in x]
y3=[i**2 for i in x]
y4=[i**3 for i in x]

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title("plot1 - sqrt")

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title("plot2 - double")

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title("plot3 - square")

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title("plot4 - cube")

plt.tight_layout()
```
