```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips= sns.load_dataset("tips")
print(tips.head())   ###Starting 5 values as in pandas
```

```python
sns.set_theme()


sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size"
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    kind="line",
    y="tip",
    hue="smoker",
    size="size"
)
```

```python
#### LIne Plot
flight=sns.load_dataset("flights")
sns.lineplot(
    data=flight,
    x="year",
    y="passengers"
)
```
