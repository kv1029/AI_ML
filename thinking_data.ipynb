```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
        
    return data
```

```python
data=load_data("data_store.json")
print(data)
```

```python
#clean and structure the data
def clean_data(data):
    cleaned_data=[]
    unique_users=set()
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        
        #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating

        #missing ages in data
        raw_age= user.get("age")
        if (raw_age==None):
            user["age"]= None

        # application
        if (user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
        
    return cleaned_data
        
```

```python

data=clean_data(data)
```

```python
#getting meaningfull insigths

def get_insights(data2):
    tot_rating=0;
    for  user in data2:
        tot_rating+=float(user["rating"])
    print(f"User rating average is {tot_rating / len(data2)}") 

    poor_ratings=0
    for user in data2:
        if (float(user["rating"])<3):
            poor_ratings+=1
    print(f"User rating average is {poor_ratings / len(data2) *100}%") 

        
```

```python
get_insights(data)
```

```python
#build recommendation feature
def get_recom(data2):
    recom=[]
    for user in data2:
        curr={}
        if (float(user["rating"])>=4):
            curr["brand"]= "Apple"
        else:
            curr["brand"]="Samsung"
        recom.append(curr)
    return recom
            
```

```python
get_recom(data)
```
